#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
with plt.style.context(('bmh')):

    fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(10,5))

    axes[0].plot(x, x**2,x, np.exp(x))
    axes[0].set_title("A")

    axes[1].plot(x, x**2,x, np.exp(x))
    axes[1].set_yscale("log")
    axes[1].set_title("B")
```

Box Whisker Plot

Histogram

Please provide a histogram of the Fuel Barrels/Year

```python
with plt.style.context(('classic')):
    fuel_barr_y = vehicles['Fuel Barrels/Year'].hist(bins=100)
    fuel_barr_y
```

Bar Chart

Please provide a bar chart of the Fuel Type on the X axis and "City MPG" on the Y axis

```python
with plt.style.context(('seaborn-notebook')):
    fuel_city_mpg = plt.bar(vehicles['Fuel Type'], vehicles['City MPG'])
    plt.xticks(rotation=90)
    plt.ylabel("Fuel type")
    plt.xlabel("City MPG")
    fuel_city_mpg
```
